```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
image_path = 'E:/thigiacmaytinh/images/R000.jpg'
image = Image.open(image_path)
gray_image = image.convert('L')
image_a = np.array(gray_image)
blurred_image_a = ndimage.gaussian_filter(image_a, sigma=1, mode="constant")
blurred_image_b = ndimage.gaussian_filter(image_a, sigma=5, mode="constant")
blurred_image_c = ndimage.gaussian_filter(image_a, sigma=9, mode="constant")
fig, axs = plt.subplots(2,2, figsize=(10, 8))

plt.subplot(2,2,1)
plt.imshow(gray_image, cmap='gray')
plt.title('Ảnh xám gốc')

plt.subplot(2,2,2)
plt.imshow(blurred_image_a, cmap='gray')
plt.title('sigma=1')

plt.subplot(2,2,3)
plt.imshow(blurred_image_b, cmap='gray')
plt.title('sigma=5')

plt.subplot(2,2,4)
plt.imshow(blurred_image_c, cmap='gray')
plt.title('sigma=9')
```
